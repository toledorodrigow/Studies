Vector Algebra can be studied by three perspectives: **Analitically, geometrically and axiomatically**. In the geometric approach, the vectors are represented as direct segments or arrows. Algebric operations (e.g. addition, subtraction, multiplication by real numbers) are defined. In the analytical approach, the vectors are fully described by numbers, called components. The analytic description arrives from the geometrical naturally.

In the axiomatic approach, vectors and vectors operations are thought of as **undefined** concepts of which we know **nothing except** that they satisfy a certain set of axioms. An algebric system, with the appropriate axioms, is called Zinear space or a linear vector space. **Therefore, the geometrical or analytical are merely two examples of linear spaces**.



# 1. Vector Algebra

The idea of using a number to locate a point in a line was known to the ancient civilizations. Descartes extended this idea by using a pair ($a_1, a_2$) and a triple ($a_1, a_2, a_3$) to locate a point in the plane and space, respectively. However, there is no need to stop with three numbers. One can, more generally, consider a *n-tuple* of real numbers:
$$(a_1, a_2, \dots, a_n)$$
for **any** integer $n>1$. Such n-tuple is called a **n-dimensionalpoint** (i.e., because any vector represent a number in the n-dimension) or an **n-dimensionalvector**, the individual numbers $a_1,a_2,\dots,a_n$ are referred to as coordinates or components of the vector. The collection of all n-dimensional vectors is called the vector space of n-tuples, or simply the n-space (denoted as $V_n$). For $n>=1$ (Only for $n=1, 2, 3$), the geometric pictures used for motivate and ilustrate the vector concepts are not available. Therefore, the study of vector algebra in higher-dimensional spaces must proceed entirely by **analytic means**.

## 1.1 Basic operation and properties

The basic operation between vectors and scalars are: *equality, addition and multiplication by scalars*.

* Equality: When two vectors A = ($a_1, a_2, \dots, a_n$) and B = ($b_1, b_2, \dots, b_n$) in $V_n$ are equal, all the correspondent components (i.e., we need to analise each dimension) is equal. Therefore, $A=B$ is the same as saying the $n$ scalar equations:
$$a_1 = b_1, \; a_2 = b_2, \;\dots, \;a_n = b_n$$
* Addition: The sum A+B is the vector obtained by adding the corresponding components
$$A+B = (a_1+b_1, a_2+b_2, \dots, a_n+b_n)$$
* Multiplication by a scalar: if $c$ is a scalar, then, $cA$ is the vector obtained multiplying each components of A by c.
$$cA = (ca_1, ca_2, \dots, ca_n)$$


```python
# Testing the operations with symbolic math
import sympy as sym
# Defining the components and the scalar
a1, a2, a3, b1, b2, b3, c = sym.symbols('a_1 a_2 a_3 b_1 b_2 b_3 c')
# Defining the vectors
A = sym.Array([a1, a2, a3]); B = sym.Array([b1, b2, b3])
# Equality
print('Equality')
sym.pprint(sym.Eq(A,B))
# Addition
print('Addition')
sym.pprint(A+B)
# Multiplication by a scalar
print('Multiplication by a scalar')
sym.pprint(c*A)
```

From these operations, its straight forward to see that **addition** is *commutative* $A+B = B+A$ (the order does not matter) and *associative* $A+(B+C) = (A+B)+C$ (Make the addition by groups does not matter). Multiplication by a scalar is *associative* $c(dA) = (cd)A$ and satisfies two distributive laws $c(A+B) = cA+cB$, $(c+d)A = cA+cD$

```python
# Defining the components from the C vector
c1, c2, c3, d = sym.symbols('c_1 c_2 c_3 d')
C = sym.Array([c1, c2, c3])
# Commutative property
commutative = sym.Eq(A+B, B+A)
print('Commutative:', commutative)
# Associative property
associative = sym.Eq(A+(B+C), (A+B)+C)
print('Associative', associative)
# Multiplication by scalars
mult_scalar = sym.Eq(c*(d*A), (c*d)*A)
print('multiplication by scalar', mult_scalar)
```

## 1.3 The dot product

The dot product (also called inner or scalar product) between two vectors $A = (a_1,a_2,\dots,a_n)$ and $B = (b_1,b_2,\dots,b_n)$ in the $V_n$ is given by the equation:

$$A\bullet B=\sum_{k=1}^{n}a_kb_k$$

Therefore, we need to multiply the correspondent components and sum everything up. The result is a scalar.

### 1.3.1 Properties

There are 4 important properties related to the inner product. Its **commutative, follow the distributive law, positive and homogene**.

* Commutative: $A\bullet B$ = $B\bullet A$
This is the same as
$$\sum_{k=1}^{n}a_kb_k = \sum_{k=1}^{n}b_ka_k\longrightarrow a_k = b_k$$
* Distributive law: $A\bullet(B+C) = A\bullet B + A \bullet C$
From the right side we have:
$$\sum_{k=1}^{n}a_k(b_k+c_k) = \sum_{k=1}^{n}a_kb_k + \sum_{k=1}^{n}a_kc_k = A\bullet B + A \bullet C$$
* Homogeneity: $c(A\bullet B) = (cA)\bullet B$
$$c\sum_{k=1}^n a_kb_k =\sum_{k=1}^n (ca_k)b_k$$
* Positivity: $A\bullet A \geq 0$
This is true because
$$\sum_{k=1}^{n}a_k^2 \longrightarrow a_k^2\geq 0$$
The equality must happen when all the a_k terms are equal to zero, therefore, $A = 0$

### 1.3.2 Cauchy-Schwarz inequality

If $A$ and $B$ are vectors in $V_n$, the following is always true:
$$(A\bullet B)^2\leq (A\bullet A)(B\bullet B)$$

In terms of components we have:

$$\sum_{k=1}^{n} a_k b_k \leq \sum_{k=1}^{n} a_k^2\sum_{k=1}^{n}b_k^2$$

* Algebric proof

Considering the following series:

$$\sum_{k=1}^{n} (a_kx+b_k)^2\geq 0 \longrightarrow \sum_{k=1}^{n} \left[(a_kx)^2 + 2xb_ka_k+b_k^2\right]\geq 0$$

Being $A=\sum_{k=1}^{n} (a_k)^2$, $B=\sum_{k=1}^{n} a_kb_k$, $C =\sum_{k=1}^{n} (b_k)^2$.

Therefore, one can rewrite the inequality as follows:

$$Ax^2+2xB+C\geq 0 \longrightarrow x^2+2B/Ax\geq -C/A$$

We know that: $(x+d)^2 = x^2+2dx+d^2$. Therefore:

$$(x+B/A)^2+C/A-B^2/A^2\geq 0$$

Therefore we have that the minimum happens when $x = -B/A$ and:
$$AC\geq B^2$$

* Vectors proof:

Considering A, B and C vectors in $V_n$, with $C = xA-yB$ and $x = B\bullet B$, $y = A \bullet B$. From the positivity property, we know that: $C\bullet C \geq 0$. Then

$(xA-yB)^2\geq 0\longrightarrow (B\bullet B)^2(A\bullet A) - 2(B\bullet B)(A\bullet B)^2+(A\bullet B)^2(B\bullet B) \geq 0$

Knowing that $B\bullet B\geq 0$ we can divide both sides by it. Therefore

$$(B\bullet B)(A\bullet A) - 2(A\bullet B)^2+(A\bullet B)^2\geq 0 
\longrightarrow (B\bullet B)(A\bullet A)\geq (A\bullet B)^2$$

The equality holds only $C=0$, given the property. Therefore, this happens only when $xA = yB \longrightarrow A = y/x B$, the vectors are parallel.

## 1.4 The Length or norm of a vector

The calculus of the length of a vector in the 2-space ($V_2$, e.g., $A = (a_1, a_2)$) or in the 3-space ($V_3$, e.g., $A = (a_1, a_2, a_3)$) follows the theorem of Pythagoras, that is:



$$\text{length of A} = ||A|| = \left(a_1^2+a_2^2\right)^{(1/2)}\longrightarrow ||A||^2 = \left(a_1^2+a_2^2\right)$$

The same pattern can be deduced for $V_3$, where



$$||A||^2 = ||A'||^2+a_3^2 = a_1^2+a_2^2+a_3^2$$

Therefore, for $V_n$, the pattern keeps repeating:

$$||A||^2 = a_1^2+a_2^2+\dots+a_n^2=\sum_{k=1}^{n}a_n^2=A\bullet A$$

Then, the length or norm of a vector is the positive square-root of the inner product from the vector to itself:

$$||A|| = \left(A\bullet A\right)^{1/2}$$

### 1.4.1 Properties

The length or norm of a vector has straightforward properties: *positivity, homogeneity*.
* Positivity: $||A|| \geq 0$, because $A\bullet A \geq 0$, and $||A|| = 0$ when $A=0$
* Homogeneity: $||cA|| = |c|||A||$. We need to remember $\sqrt{c^2} = |c|$, this is the positive square-root.
We got that:
$$||cA|| = \left(cA\bullet cA\right)^{1/2} = \left(c^2A\bullet A\right)^{1/2} = |c|\left(A\bullet A\right)^{1/2}=|c|||A||$$

### 1.4.2 Cauchy-Schwarz and triangle inequalities

We know that the Cauchy-Schwarz inequality is: $(A\bullet B)^2\leq (A\bullet A)(B\bullet B)$. Therefore, knowing that the length of a vector squared is $||A||^2 = A\bullet A$, the inequality can be rewrote to:

$$(A\bullet B)^2\leq ||A||^2||B||^2\longrightarrow A\bullet B \leq ||A||||B||$$

From the parallelogram rule, we know that a vector $C = A+B$



The triangle inequality states that any side of the triangle must be last or equal to the sum of the other two sides. Therefore:

$$||C|| \leq ||A|| + ||B||\longrightarrow ||A+B|| \leq ||A|| + ||B||$$

To prove, we need to square both sides:

$$||A+B||^2 \leq (||A||+ ||B||)^2\longrightarrow C\bullet C \leq ||A||^2+2||A||||B||+||B||^2$$

We know that: 

$$C\bullet C = (A+B)\bullet(A+B) = A\bullet A + 2A\bullet B + B\bullet B = ||A||^2+2A\bullet B + ||B||^2$$

Then:
$||A||^2+2A\bullet B + ||B||^2 \leq ||A||^2+2||A||||B||+||B||^2 \longrightarrow A\bullet B \leq ||A||||B||$

In which:

$$A \bullet B \leq ||A||||B|| \longrightarrow (A\bullet B)^2 \leq (A\bullet A) (B\bullet B)$$

Is the Cauchy-Schwarz inequality.

## 1.5 Orthogonality of vectors

From the parallelogram rule, we know that:



$C = A+B\longrightarrow ||C||^2 = ||A+B||^2$

Then, for any sum of two vectors in $V_n$:

$C\bullet C = (A+B)\bullet(A+B) = (A\bullet A) + 2A\bullet B + B\bullet B = ||A||^2+2A\bullet B + ||B||^2$

For the extreme case of the rectangular triangle:



We know by the Pythagorean identity: $||C||^2 = ||A||^2+||B||^2$

Therefore, comparing the expression with the general one, for a vector formed by two *perpendicular* vectors (*orthogonals*), the inner product is:

$$A\bullet B = 0$$

## 1.6 Projections and angle between vectors in n-space

Consider three vectors (A, B and C) in $V_2$, in which $A=tB+C$, this is, $tB$ is the projection of $A$ in $B$ and $C\bullet B = 0$ (they are perpendicular or orthogonal). 



Therefore, taking the dot product of each element by $B$ we have:
$$A\bullet B = tB\bullet B + C\bullet B = tB\bullet B\longrightarrow t=\frac{A\bullet B}{B\bullet B} =\frac{A\bullet B}{||B||^2}$$
From the trigonometric relations, we have that:
$$||tB|| = \cos(\theta)||A||\longrightarrow \cos(\theta) = t \frac{||B||}{||A||}=\frac{A\bullet B}{||A||||B||}$$
Then, we end with:
$$A\bullet B = ||A||||B||\cos(\theta)$$
From the Cauchy-Schwarz inequality: $(A\bullet B)^2\leq ||A||^2||B||^2\longrightarrow |A\bullet B| \leq ||A||||B||\longrightarrow \frac{|A\bullet B|}{||A||||B||}\leq 1$. Therefore:

$$-1\leq\frac{A\bullet B}{||A||||B||}\leq 1$$

This concept of angle can be defined in all $V_n$, where $\theta$ is:

$$\theta =\arccos{\frac{A\bullet B}{||A||||B||}}$$
with $0\leq\theta \leq \pi$


```python
import matplotlib.pyplot as plt
import numpy as np

cos_theta = np.linspace(-1,1,1000);
theta = np.arccos(cos_theta)
theta_multiples = np.linspace(0,2*np.pi,1000)
fig,(ax1,ax2)= plt.subplots(2,1)
ax1.plot(theta_multiples, np.cos(theta_multiples))
ax1.plot(theta, cos_theta);
ax2.plot(np.cos(theta_multiples), theta_multiples, cos_theta, theta);
```

## 1.7 The unit coordinate vectors

A vector $X=(x_1, x_2, \dots, x_n)$ in $V_n$ can be described by a finite sum of the components multiplied by a unit vector. That is:

$$X = x_1(1,0,\dots,0)+x_2(0,1,\dots,0)+\dots+x_n(0,\dots,1) = x_1E_1+x_2E_2+\dots+x_nE_n = \sum_{k=1}^{n}x_kE_k$$

In which $E_n$ is the nth unit coordinate vector. The norm of any unit vector is $||E_n|| = (0^2+\dots+1^2)^{1/2} = 1$. And any two unit coordinate vectors are orthogonal to each other, and all are mutually orthogonal $E_k\bullet E_j = (0,\dots,1,\dots,0)\bullet(0,\dots,\dots,1,0) = 0*0+\dots+1*0+0*1+0*0+\dots$

Therefore, the sum $\sum_{k=1}^{n}x_kE_k$ is called a *linear combination* of vectors $E_1,\dots,E_k$ and every vector in $V_n$ can be expressed (and uniquely) as a linear combination of the unit vectors(i.e., the unit coordinate vectors $E_1,\dots,E_k$ span, or can represent all the linear set $V_n$). In $V_2$, the unit vectors are represented as $i=\hat{a}_x=(1,0),j=\hat{a}_y=(0,1)$. In $V_3$, the unit vectors are represented as $i=\hat{a}_x=(1,0,0),j=\hat{a}_y=(0,1,0), k=\hat{a}_z = (0,0,1)$.



Therefore, any algebric manipulation between vectors can treat the vectors as finite sums $\sum_{k=1}^{n}x_kE_k$. For example, the addition between two vectors $A=\sum_{k=1}^{n}a_kE_k$ and $B=\sum_{k=1}^{n}b_kE_k$ is given by:

$$A+B = \sum_{k=1}^{n}(a_k+b_k)E_k$$

## 1.8 The linear span of a finite set of vectors

If $S = \{A_1,\dots,A_k\}$ is a set of $k$ vectors in $V_n$, with $k$ being the number of vectors ($k<n\quad \text{or} \quad k=n \quad\text{or} \quad k>n$), we say that $S$ span a vector $X = (x_1, x_2, \dots, x_n)$ if $X$ can be expressed as a linear combination of $S$ vectors (it must be all the vectors in $S$).

$$X = \sum_{i=1}^{k} c_iA_i$$

$c_i$ can be different than $x_i$, i.e., $c_i$ will be equal to $x_i$ if $S$ is the unit coordinate set of vectors. This is when $S = \{E_1 = (1,0,\dots,0),E_2 = (0,1,\dots,0),\dots, E_k = (0,\dots,1)\}$.

The set of all vectors spanned by $S$ (all the possible linear combinations of $S$) is denoted by $L(S)$. If $L(S) = V_n$, $S$ spans the whole space.

Every set $S=\{A_1, \dots, A_k\}$ spans the zero vector since $0 = 0A_1+0A_2+\dots+0A_k$ (trivial solution when all the coeficients are zero: $c_1=c_2=\dots=c_k=0$). However, if $A_2 = 2A_1$ (one vector of S is parallel - scalar multiple of another), there are many nontrivial solutions for express zero:

$$0 = \sum_{i=1}^k c_iA_i = 2tA_1 - tA_2 + 0A_3 + \dots + 0A_k$$
with $c_1 = 2t$ and $c_2 = -t$. for any $t\neq0$.

Nonetheless, **we are interested in sets $S$ that span vectors in exactly one way**. A set $S$ span $X$ uniquely if $S$ spans $X$ and:

$$X = \sum_{i=1}^k c_iA_i \quad \text{and} \quad X = \sum_{i=1}^{k} d_iA_i \quad \text{implies} \quad c_i=d_i \forall i$$

And **if a set $S$ spans the zero vector uniquely, every vector in $L(S)$ is spanned uniquely**.
* proof: considering that if $S$ spans every vector in $L(S)$ uniquely, then it certanly spans the 0 uniquely. Therefore:
$$X-X = 0 = \sum_{i=1}^k (c_i-d_i)A_i$$
But since $S$ spans 0 uniquely, we must have $c_i-d_i=0$ for all $i$.


## 1.9 Linear Independence

Any set $S=\{A_1, A_2, \dots, A_n\}$ can be denoted in two classifications: linearly *independent* or linearly *dependent*. To a set be independent, it must span the zero vector uniquely (i.e., only with the trivial representation). Therefore (supposing a zero vector $X=0$):

$$X = \sum_{i=1}^{k}c_iA_i = 0, \quad \text{implies that every} \quad c_i = 0$$

Therefore, to be dependent, the set $S$ must span the zero vector in a nontrivial way

$$X = \sum_{i=1}^{k}c_iA_i = 0, \quad \text{implies that not all} \;c_i\;\text{is zero}$$

### 1.9.1 Sets and subsets

Logically, we can define for a subset $T$ of $S$ that:



* If $T$ is dependent, then, $S$ is also dependent.
* If $S$ is independent, then, $T$ is also independent.
* If $T$ is independent, then $S$ can be either dependent or independent.

Suppose a subset of vectors $T = \{i,j\}$ of $S=\{i,j,i+j\}={(1,0),(0,1),(1,1)}$ in $V_2$. For $T$ we got:

$$\sum_{i=1}^{k}c_iA_i = 0 \longrightarrow c_1(1,0)+c_2(0,1) = 0, \quad c_1=c_2=0$$
Then, $T$ is a linearly independent set. However, for $S$:
$$\sum_{i=1}^{k}c_iA_i = 0 \longrightarrow c_1(1,0)+c_2(0,1)+c_3(1,1)=0, \quad c_1 = c_2 = -c_3$$
Therefore, $S$ is a linearly dependent set (does not have only the trivial solution).

The third vector, $i+j$, is in the linear span of $T$. If we adjoint to $i$ and $j$ any vector in the linear span of $T$, we get a dependent set. Therefore, if $T$ or $S = \{A_1,\dots,A_k\}$ is a linearly independet set of k vectors in $V_n$, and $L(S)$ the linear span of $S$, then, every set of $k+1$ vectors in $L(S)$ is linearly dependent (this is exactly what happend when we add (i+j) to $T$). In other words,**if we add a vector in the linear span L(S) of a linearly independent set $S$ to this set, it becomes linearly dependent**.

### 1.9.2 Orthogonal Sets

For two vectors be orthogonal, the inner product from the vectors must be 0, i.e., $A_1\bullet A_2=0$. For a set $S = {A_1, \dots, A_k}$ be orthogonal, each pair of vectors must be orthogonal, i.e., $A_i\bullet A_j = 0$ whenever $i\neq j$. Therefore, **each two vectors in a orthogonal set are perpendicular**.

**Any orthogonal set $S=\{A_1, \dots, A_k\}$ is linearly independent**. And if $S$ spans a vector $X=\sum_{i=1}^{k}c_iA_i$, the coefficient terms are:

$$c_i = \frac{X\bullet A_i}{A_i\bullet A_i} = \frac{X\bullet A_i}{||A_i||^2}$$

* Proof: Being

$$X = 0 = \sum_{i=1}^{k} c_iA_i = c_1A_1+\dots+c_kA_k$$

We can take the dot product of $X$ with respect to $A_1$:

$$0 = c_1A_1\bullet A_1+\dots+c_kA_k\bullet A_1 = X\bullet A_1 = c_1A_1\bullet A_1$$

Considering that in a orthogonal set $A_i\bullet A_j = 0$. Because $A_1\neq 0$, $A_1\bullet A_1\neq 0$, then, the only solution is $c_1=0$. We can extend this for any $c_k=0$, then $c_1=\dots=c_k = 0$ is the trivial solution. **Therefore, $S$ is linearly independent.**

If $X = \sum_{i=1}^{k} c_iA_i$, we can take the dot product of $X$ with respect to $A_1$:

$$X\bullet A_1 = c_1A_1\bullet A_1+\dots+c_kA_k\bullet A_1 = X\bullet A_1 = c_1A_1\bullet A_1$$

Then, we can generalize the expression for:

$$X\bullet A_k = c_kA_k\bullet A_k\longrightarrow c_j = \frac{X\bullet A_j}{A_j\bullet A_j} =\frac{X\bullet A_j}{||A_j||^2}$$

If all the vectors in $S = {A_1, \dots, A_k}$ has norm 1 (i.e., ||A_j|| = 1), **the set is called an orthonormal set** and the expression for $c_j$ is simplified to:

$$c_j = X\bullet A_j$$

## 1.10 Bases

A set $S = \{A_1, \dots, A_k\}$ of vectors in $V_n$ is called a basis for $V_n$ if $S$ spans every vector in $V_n$ uniquely. Therefore, **a basis is a linearly independent set that spans the whole space $V_n$**. The set of unit coordinate vectors is a orthogonal basis $S=\{E_1, \dots, E_k\}$.

For a given vector space $V_n$, the basis have the following properties:
* 1) Every basis contains the exactly $n=\text{dimension of the vector space}$ vectors.

proof: Let $S$ and $T$ be two bases with $k$ and $r$ vectors, respectively. if $r>k$, then $T$ contain at least $k+1$ vectors in $L(T)$, what makes T linearly dependent. **Then neither $r>k$ or $k>r$ are possible for basis**.

* 2) Any set of linearly independent vectors is a subset of some basis.

The basis has $n$ vectors, a subset can have last than $n$

* 3) Any set of n linearly independent vectors is a basis for $V_n$.

## 1.11 The vector space $V_n(C)$ of n-tuples of complex numbers

The vector space $V_n$ was defined as the collection of n-tuples of real numbers (e.g., $A = (a_1, \dots, a_n)$ is a vector of $V_n$). If all the scalars are replaced by complex numbers, the new algebric system so obtained is called *complex vector space* and is denoted by $V_n(C)$. Since complex numbers satisfy the same field properties as real numbers (i.e., operations involving addition and multiplication) all theorems about real vector space $V_n$ that uses the field properties of the real numbers are also valide for $V_n(C)$. For example:

$$A = B\quad \text{means} \quad a_i = b_i \forall i$$
$$A+B = (a_1+b_1, \dots, a_n+b_n), \qquad cA = (ca_1, \dots, ca_n)$$

Complex vector spaces arise naturally in the theory of linear differential equations and electrical engineering problems, therefore, this generalization is of considerable importance.

Nevertheless, in theorems that involve dot products, some small changes must be made. The dot product for a real vector space $V_n$ is given by:

$$A\bullet B = \sum_{k=1}^{n} a_kb_k = \text{real number}$$

When we take the dot product from the same vector, we have:

$$A\bullet A = \sum_{k=1}^{n} a_k^2 \geq 0, \quad \text{because} \quad a_k^2\geq 0, \quad \text{for real numbers}$$

However, considering complex numbers, the dot product $A\bullet A = \sum_{k=1}^{n} a_k^2$ can be nonpositive. Therefore, for this sake, we use the *conjugate*. We know that, for any complex number $A = a_r+a_ij$:

$$A\bar{A} = (a_r+a_ij)(a_r-a_ij) = a_r^2+a_i^2 = |A|^2\geq 0 \quad\text{for all complex number}$$

Therefore, for consider complex numbers, the dot product must have the following change:

$$A\bullet B = \sum_{k=1}^{n} a_k\bar{b}_k$$

### 1.11.1 Properties

Given the new dot product formulation, the some properties are changed:

* (a) Commutativity: $A\bullet B = \bar{B}\bullet \bar{A}$

proof: $A\bullet B = \sum_{k=1}^{n}a_k\bar{b}_k = \sum_{k=1}^{n}\bar{b}_ka_k = \sum_{k=1}^{n}\bar{b}_k\bar{\bar{a_k}} = \bar{B}\bullet \bar{A}$

* (b) Associativity: $A\bullet(B+C) = A\bullet B + A\bullet C$, $(B+C)\bullet A= B\bullet A + C\bullet A$

proof: $A\bullet (B+C) = \sum_{k=1}^{n}a_k(\bar{b}_k+\bar{c}_k) = \sum_{k=1}^{n}a_k\bar{b}_k + \sum_{k=1}^{n}a_k\bar{c}_k = A\bullet B + A\bullet C$

* (c) Commutativity when multiplied by a constant: $c(A\bullet B) = (cA)\bullet B = A\bullet (\bar{c}B)$

proof: $cA\bullet B = \sum_{k=1}^{n}(ca_k)\bar{b}_k = \sum_{k=1}^{n}a_k\overline{(\bar{c}b_k)} = (cA)\bullet B = A\bullet (\bar{c}B)$

* (d) Positivity: $A\bullet A \geq 0$

proof: $A\bullet A = \sum_{k=1}^{n}a_k\bar{a}_k = \sum_{k=1}^{n}|a_k|^2 \geq 0$

### 1.11.2 Exercises

```python
# Defining vectors
A = np.array([1, 1j]); B = np.array([1j,-1j]); C = np.array([2j,1])
#A) A.B
def inner_product(v1,v2):
    inner = 0
    for k in range(0,len(v2)):
        inner+=v1[k]*np.conjugate(v2[k])
    return inner
print('a) A.B', inner_product(A,B));print('b) B.A', inner_product(B,A));print('c) (iA).B', inner_product(1j*A,B))
print('d) A.(iB)', inner_product(A,1j*B));print('e) (iA).(iB)', inner_product(1j*A,1j*B));print('f) B.C', inner_product(B,C))
print('g) A.C', inner_product(A,C));print('h) (B+C).A', inner_product(B+C,A));print('i) (A-C).B', inner_product(A-C,B))
print('f) (A-iB)(A+iB)', inner_product(A-1j*B,A+1j*B))
```
